<h3>Handling the missing values</h3>

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

CompetitionDistance — 2186 nulls<br>
No nearby competitor, so fill with large distance value

```python
df['CompetitionDistance'] = df['CompetitionDistance'].fillna(200000)
```

CompetitionOpenSinceMonth & Year — 268619 nulls<br>
Store has no competitor, so competition "opened" at time 0

```python
df['CompetitionOpenSinceMonth'] = df['CompetitionOpenSinceMonth'].fillna(0)
df['CompetitionOpenSinceYear'] = df['CompetitionOpenSinceYear'].fillna(0)
```

Promo2SinceWeek, Promo2SinceYear, PromoInterval — 423307 nulls<br>
Store is not participating in Promo2, so fill with 0 / 'None'

```python
df['Promo2SinceWeek'] = df['Promo2SinceWeek'].fillna(0)
df['Promo2SinceYear'] = df['Promo2SinceYear'].fillna(0)
df['PromoInterval'] = df['PromoInterval'].fillna('None')
```

```python
df.isnull().sum()
```

<h3>Fixing data types and extracting date features</h3>

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Year']      = df['Date'].dt.year
df['Month']     = df['Date'].dt.month
df['Week']      = df['Date'].dt.isocalendar().week.astype(int)
df['DayOfWeek'] = df['Date'].dt.dayofweek                                 # 0=Monday, 6=Sunday
df['Quarter']   = df['Date'].dt.quarter
```

Encoding categorical columns as category dtype

```python
df['StoreType']   = df['StoreType'].astype('category')
df['Assortment']  = df['Assortment'].astype('category')
df['StateHoliday'] = df['StateHoliday'].astype(str)
```

Verifying the data ranges

```python
print("Date range:", df['Date'].min(), "to", df['Date'].max())
print("Unique years:", df['Year'].unique())
```

<h3>Feature engineering</h3>

Log sales — handles right-skewed distribution

```python
df['log_sales'] = np.log1p(df['Sales'])
```

Competition proximity flag

```python
df['competition_flag'] = (df['CompetitionDistance'] < 1000).astype(int)
```

Holiday flag — any kind of holiday active

```python
df['is_holiday'] = (
    (df['StateHoliday'] != '0') |
    (df['SchoolHoliday'] == 1)
).astype(int)
```

Promo2 active flag — is Promo2 currently running this month? <br>
PromoInterval tells you which months Promo2 runs (e.g. "Jan,Apr,Jul,Oct")

```python
month_map = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr',
             5:'May', 6:'Jun', 7:'Jul', 8:'Aug',
             9:'Sep', 10:'Oct', 11:'Nov', 12:'Dec'}
```

```python
def promo2_active(row):
    if row['PromoInterval'] == 'None':
        return 0
    months = row['PromoInterval'].split(',')
    return 1 if month_map.get(row['Month'], '') in months else 0
```

```python
df['promo2_active'] = df.apply(promo2_active, axis=1)
```

Weekend flag

```python
# Weekend flag
df['is_weekend'] = (df['DayOfWeek'] >= 5).astype(int)
```

```python
print("New columns added:")
print(['log_sales', 'competition_flag', 'is_holiday',
       'promo2_active', 'is_weekend'])
print(df[['Sales', 'log_sales', 'Promo', 'promo2_active',
          'is_holiday', 'competition_flag']].head(10))
```

<h3>Sanity or Data Validation checks before moving on</h3>

Check 1: Sales distribution check — should be right-skewed, no negatives

```python
print("Sales stats:")
print(df['Sales'].describe())
print("Negative sales:", (df['Sales'] < 0).sum())  # Should be 0
```

Check 2: Suspicious Stores

It checks for each store, what percentage of days had zero sales?

```python
store_zero_pct = df.groupby('Store').apply(lambda x: (x['Sales'] == 0).mean()).sort_values(ascending=False)
suspicious_stores = store_zero_pct[store_zero_pct > 0.3]
```

```python
print(f"\nSuspicious stores (>30% zero sales when open): "
      f"{len(suspicious_stores)}")
print(suspicious_stores.head(10))
```

Check 3: Promo coverage — what % of open days had a promotion?

```python
print(f"\nPromo coverage: {df['Promo'].mean():.1%}")
print(f"Promo2 active coverage: {df['promo2_active'].mean():.1%}")
```

Check 4: Store count per type

```python
print("\nStores per type:")
print(df.groupby('StoreType')['Store'].nunique())
```

Check 5: Date continuity — are there gaps?

```python
date_range = pd.date_range(df['Date'].min(), df['Date'].max(), freq='D')
actual_dates = df['Date'].unique()
print(f"\nExpected date range length: {len(date_range)}")
print(f"Actual unique dates in data: {len(actual_dates)}")
```
